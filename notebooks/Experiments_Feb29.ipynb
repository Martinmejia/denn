## Log Spacing (`np.geomspace`): 10K iters, 1K pts

```python
import numpy as np
```

```python
lhs = np.geomspace(0.02, 1, num=500) - 1.02
rhs = np.flip(-lhs, axis=0)
geomgrid = np.concatenate([lhs, rhs])
```

```python
import matplotlib.pyplot as plt
plt.hist(geomgrid, density=True);
```
