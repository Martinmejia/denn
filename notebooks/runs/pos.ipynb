```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def anal_sol(x, y):
    return np.sin(np.pi*y) * np.sinh(np.pi*(1-x)) / np.sinh(np.pi)
```

```python
x = np.linspace(0, 1, 30)
y = np.linspace(0, 1, 30)
xx, yy = np.meshgrid(x, y)
```

```python
anal = anal_sol(xx, yy)
```

```python
anal.shape
```

```python
plt.contourf(xx, yy, anal, cmap="Reds")
```
