```python
from denn.utils import Discriminator, Generator
from denn.sho.gan_sho import train_GAN_SHO_unsupervised, train_GAN_SHO_semisupervised
import torch.nn as nn
```

# Unsupervised

```python
G = Generator(in_dim=1, out_dim=1,
              n_hidden_units=64,
              n_hidden_layers=8,
              activation=nn.Tanh(), # twice diff'able activation
              output_tan=True,      # true output range should be (-1,1) if True
              residual=True)

D = Discriminator(in_dim=2, out_dim=1,
                  n_hidden_units=64,
                  n_hidden_layers=8,
                  activation=nn.Tanh(),
                  unbounded=True, # true for WGAN
                  residual=True)

res = train_GAN_SHO_unsupervised(G, D, d_lr=2e-4, final_plot=True, num_epochs=10000)
```

# Semi-supervised

```python
G = Generator(in_dim=1, out_dim=1,
              n_hidden_units=64,
              n_hidden_layers=8,
              activation=nn.Tanh(), # twice diff'able activation
              output_tan=True,      # true output range should be (-1,1) if True
              residual=True)

D = Discriminator(in_dim=2, out_dim=1,
                  n_hidden_units=64,
                  n_hidden_layers=8,
                  activation=nn.Tanh(),
                  unbounded=True, # true for WGAN
                  residual=True)

D2 = Discriminator(in_dim=2, out_dim=1,
                  n_hidden_units=64,
                  n_hidden_layers=8,
                  activation=nn.Tanh(),
                  unbounded=True, # true for WGAN
                  residual=True)

res = train_GAN_SHO_semisupervised(G, D, D2, d_lr=2e-4, final_plot=True, num_epochs=10000)
```
