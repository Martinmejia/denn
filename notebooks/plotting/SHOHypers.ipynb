```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../../experiments/csvs/sho_GAN_hypertune_biggest_ever.csv', index_col=0)
df.head(5)
```

```python
from pandas.io.json import json_normalize
```

```python
res = pd.concat((df.mse, json_normalize(df.hypers.apply(eval))), axis=1)
```

```python
top_n = 10
top_2k = res.sort_values(by='mse').loc[res.gan_niters==2000].iloc[0:top_n]
top_5k = res.sort_values(by='mse').loc[res.gan_niters==5000].iloc[0:top_n]
top_10k = res.sort_values(by='mse').loc[res.gan_niters==10000].iloc[0:top_n]
```

```python
merge1 = pd.merge(top_2k, top_5k, on=list(res.columns.drop(["mse", "gan_niters"])))
```

```python
merge2 = pd.merge(merge1, top_10k, on=list(res.columns.drop(["mse", "gan_niters"])))
```
