```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../../experiments/csvs/nlo_GAN_hypertune_biggest_ever.csv', index_col=0)
df.head(5)
```

```python
from pandas.io.json import json_normalize
```

```python
res = pd.concat((df.mse, json_normalize(df.hypers.apply(eval))), axis=1)
```

```python
top_n = 10
top_5k = res.sort_values(by='mse').loc[res.gan_niters==5000].iloc[0:top_n]
top_10k = res.sort_values(by='mse').loc[res.gan_niters==10000].iloc[0:top_n]
top_20k = res.sort_values(by='mse').loc[res.gan_niters==20000].iloc[0:top_n]
```

```python
merge1 = pd.merge(top_5k, top_10k, on=list(res.columns.drop(["mse", "gan_niters"])))
```

```python
merge2 = pd.merge(merge1, top_20k, on=list(res.columns.drop(["mse", "gan_niters"])))
```
