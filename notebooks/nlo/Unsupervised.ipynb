```python
from denn.utils import Discriminator, Generator
from denn.nlo.gan_nlo import train_GAN_NLO_unsupervised, train_GAN_NLO_semisupervised
import torch.nn as nn
```

# Unsupervised

```python
D = Discriminator(in_dim=2, out_dim=1,
                  n_hidden_units=32,
                  n_hidden_layers=4,
                  activation=nn.Tanh(),
                  unbounded=True,  
                  residual=True)

G = Generator(in_dim=1, out_dim=1,
              n_hidden_units=32,
              n_hidden_layers=4,
              activation=nn.Tanh(),
              output_tan=False,  
              residual=True)

res = train_GAN_NLO_unsupervised(G, D, d_lr=2e-4, num_epochs=10000, final_plot=True, realtime_plot=True, check_every=100)
```
